```python
data['sepal_length']=data['sepal_length'].fillna(data['sepal_length'].mean())
data.isnull().sum()
```

```python
data['sepal_width']=data['sepal_width'].fillna(data['sepal_width'].median())
data.isnull().sum()
```

```python
data['petal_length']=data['petal_length'].fillna(data['petal_length'].median())
data.isnull().sum()
```

```python
data.dropna()
```

```python
conditional_sepal_length=data[data['sepal_length']<=5.5]
print(conditional_sepal_length.head())
```

```python
conditional_species=data[(data['species']!='Iris Setosa')]
print(conditional_species)
```

```python

conditional_species=data[(data['species']).isin(['versicolor','virginica'])]
print(conditional_species)
```

Feature Engineering

```python
#creation of new column

data['petalarea']=data['petal_length']*data['petal_width']
data.head()
```

```python
data["petal_length Group"]=pd.cut(data['petal_length'],bins=[0,2,4,6],labels=['small','medium','large'])
data.head()
```

1. Label encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['species_encoded']=le.fit_transform(data['species'])
print(data.head)
```

Feature scaling ( standardization )

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_data=scaler.fit_transform(data.iloc[:,0:4])
data_scaled=pd.DataFrame(scaled_data,columns=data.columns[0:4])
print(data_scaled.head())
```

Feature scaling (min-max)

```python
from sklearn.preprocessing import MinMaxScaler
m=MinMaxScaler()
m_scaled=m.fit_transform(data.iloc[:,0:4])
data_minmax=pd.DataFrame(m_scaled,columns=data.columns[0:4])
print(data_minmax.head())
```
