To check the password is strong or not ( for loop )

```python
str=input()
count=0
upper=0
lower=0
num=0
special=0
for char in str:
  count+=1
  if 'A'<=char<='Z':
    upper+=1
  elif 'a'<=char<='z':
    lower+=1
  elif '0'<=char<='9':
    num+=1
  else:
    special+=1
if 8<=count<=16 and upper>=1 and lower>=1 and num>=1 and special>=1:
  print("strong password")
else:
  print("Not strong password")

```

To find and extract the square of n natural numbers. (while loop)

```python
n=int(input())
num=1
dict={}
while num!=n:
  dict[num]=num**2
  num+=1
print(dict)
```

Implement 10 combinations of data operators.

```python
a = [10, 20, 30]
b = [5, 10, 15]

add = []
sub = []
mul = []
div = []
mod = []
power = []
gt = []
lt = []
eq = []
logical_and = []

for i in range(len(a)):
    add.append(a[i] + b[i])
    sub.append(a[i] - b[i])
    mul.append(a[i] * b[i])
    div.append(a[i] / b[i])
    mod.append(a[i] % b[i])
    power.append(a[i] ** b[i])
    gt.append(a[i] > b[i])
    lt.append(a[i] < b[i])
    eq.append(a[i] == b[i])
    logical_and.append(a[i] > 10 and b[i] > 10)

print("Addition:", add)
print("Subtraction:", sub)
print("Multiplication:", mul)
print("Division:", div)
print("Modulus:", mod)
print("Power:", power)
print("Greater than:", gt)
print("Less than:", lt)
print("Equal to:", eq)
print("Logical AND:", logical_and)
```
